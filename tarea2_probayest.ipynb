```python
#7. En un grupo de 60 estudiantes se separa en 3 clases del mismo tamaño. Jack y Jill son estudiantes de este grupo ¿cual es la 
#probabilidad de que les toque en la misma clase?

## Cada clase sería de 20 estudiantes, entonces hay 3 clases de las cuales cualquiera de los 2 van a una de ellas. Ahora como uno de ellos
## ya esta en una clase, solo falta que el otro vaya a la misma, entonces solo hay tres posibilidades, por lo que es 1/3
## se intenta mediante una aproximacion de montecarlo
import random
def clases(intentos=1000000):
    misma_clase=0 #ponemos un contador

    for _ in range(intentos):
        jack= random.choice([1, 2, 3])
        jill= random.choice([1, 2, 3])

        if jack == jill:
            misma_clase += 1
        
    probabilidad_clase= misma_clase/intentos
    return probabilidad_clase
intentos=100000
probabilidad_clase=clases(intentos)
probabilidad_clase
```

```python
#8. Se tiene la moneda A y B. La moneda A tiene 1/12 probabilidad de salir aguila o sol en cada tirada. La probabilidad de sacar sol en la 
# moneda B es 1/3. Si seleccionamos una moneda aleatoriamente y la tiramos hasta obtener sol. La probabilidad de seleccionar la moneda A
# es 1/4 y la probabilidad de seleccionar B es 3/4, ¿cual es el numero esperado de veces que tiraremos la moneda para obtener sol?

##necesitamos un valor esperado. Estamos en un problema de probabilidad condicional y se usa la ley de esperanza total
##E=P(A)⋅E_A+P(B)⋅E_B

p_a=1/12
p_b=1/3
e_a=1/4
e_b=3/4

esperanza=p_a*e_a+p_b*e_b
esperanza
```

```python
#9.Una compañia de seguros vende seguro de vida que paga 240000.- pesos y la póliza es válida por 1 año. Le vende el seguro a una jóven de 
#25 años por 210 pesos. La probabilidad de que la cliente sobreviva el año es 0.999592 . ¿Cual es el valor esperado de esta
#póliza para la compañía de seguros?

##Aqui hay de 2, que sobreviva o que fallezca
## si sobrevive pues el seguro gana 210 pesos, pero si fallece hay perdida para ellos de total del seguro menos sus $210 y esa probabilidad es 1-sobrevivir, o sea el complemento
##Calcularemos el valor esperado para la poliza de seguros, por lo que podemos usar la formula anterior para resolver el problema
p_sobrevive = 0.999592
p_fallece = 1 - p_sobrevive
seguro_pago = 210
pago_fallecimiento = 240000
perdida_fallece = -(pago_fallecimiento - seguro_pago) # es negativo porque se supone que es una perdida para ellos

# Calcular el valor esperado
valor_es= (p_sobrevive * seguro_pago) + (p_fallece * perdida_fallece)
valor_es
```

```python
#13. ¿Cuál es la probabilidad de que estudiantes tengan su cumpleaños en el mismo día en una clase de 30 estudiantes?

##Si analizamos a datelle, observamos que podemos hacer el analisis definiendo la probabilidad del cumpleaños de que cualquier estudiante
## de ese grupo de 30 sea cualquier dia, o sea 1, de ahi, al siguiente estudiante ya no seria la misma, sino 364/365, pues es un 
## problema de probabilidad condicional. Y asi sucesivamente con los restantes alumnos. Si calculamos la probabilidad de que todos sean
## diferentes seria p_dif=1-364/365-363/365-...
prob_diferentes=1.0
for i in range(30):
        prob_diferentes *= (365 - i) / 365
prob_diferentes

## y si sacamos el complemento de esto, es decir, que al menos 2 estudiantes cumplan el mismo dia es justo 1-prob_diferentes
com=1-prob_diferentes
com
```

```python
#14. Si estás vendiendo sandwiches. Y el 70% de las personas elgien de huevo y el resto de pollo. 
#¿Cuál es la probabilidad de vender 2 sandwiches de huevo dentro de los 3 siguientes clientes?

##Usamos una distribucion binomial, pues necesitamos una probabilidad de exito. Donde el exito es elegir el sandwich de huevo, los exitos
## las veces que elegiran el del huevo, y las pruebas son los clientes
##P(k exitos de n intentos)=(n,k)p^k(1-p)^(n-k)
import math
huevo=0.7
sandwiches=2
clientes=3
combina1= math.comb(clientes, sandwiches)
proba_huevo= combina1 * (huevo ** sandwiches) * ((1 - huevo) ** (clientes - sandwiches))
proba_huevo
```

```python
#15. Un hospital ha registrado que el 75% de pacientes sufriendo una enfermedad mueren debido a ella. ¿Cuál es la probabilidad de que 
#4 de 6 pacientes seleccionados al azar se recuperen?

##Aqui hay de 2, o se recupera o muere. Lo que nos lleva a pensar en casos de exito y relacionarlo con una distribucion binomial.
## El 25% sería la probabilidad de que se recuperen de la enfermedad, el número de exitos serían 4 y las tomas son 6.
import math
proba_exito=0.25
exitos_pacientes=4
pacientes=6
combina2=math.comb(pacientes,exitos_pacientes)
proba_pacientes=combina2*(proba_exito**exitos_pacientes)*((1-proba_exito)**(pacientes-exitos_pacientes))
proba_pacientes
```

```python
#17. Jack tiene 2 monedas. Una normal (2 lados distintos) y otra que tiene "sol" por los 2 lados. Si no se da cuenta que moneda arrojó y 
#cayó "sol". ¿Cual es la probabilidad de que la moneda que arrojó sea la que tiene "sol" por ambos lados?

## Vamos a resolver el problema usando el teorema de bayes y definimos
#eventoA= Jack arroja la moneda de los 2 soles(de entre sus 2 monedas)
#eventoB= caer sol
##como jack tiene 2 monedad, entonces la proba_a=1/2
##la probabilidad de B cuando ya sucedio A es 1, pues si ya lanzó la moneda de dos soles, es logico que el resultado es sol
## la probabilidad de p(B|A^c) o sea, la probabilidad de B tal que el complemento de A. El complemento de A es arrojar la moneda normal y necesitamos que caiga sol, entonces es 1/2
## ahora la probabilidad de B es lo que queremos calcular, pues pudo haber lanzado cualquiera de las 2, que ya es diferente 
##El complemento de A es arrojar la moneda normal y eso es 1/2
## P(B)=P(B∣A)⋅P(A)+P(B∣A c)⋅P(A c)
p_a=1/2
p_B_A=1
p_B_Ac=1/2
p_Ac=1/2

p_B=p_B_A*p_a+p_B_Ac*p_Ac

##finalmente calculamos p(A|B)= p(B|A)*p(A)/p(B)
p_A_B= p_B_A*p_a/p_B
p_A_B
```

```python
#18. La probabilidad de tener cierta enfermedad en una población es de 0.00148. La prueba de laboratorio para dicha enfermedad tiene
#una tasa de 93% de verdaderos positivos (sale positiva cuando realmente se tiene la enfermedad el 93 % de las veces). 
#Y su tasa de verdaderos negativos es de 99 % (Sale negativa cuando no se tiene la enfermedad el 99 % de las veces). 
#¿Cuál es la probabilidad de tener la enfermedad dado que la prueba salió positiva? 
#Nota: Recordar el teorema de Bayes P(A | B)P(B) = P(B|A)P(A)

##Vamos a denotar otras variables para que no haya problemas con el anterior problema
## p(x|y)= p(y|x)*p(x)/p(y)
## eventox= tener la enfermedad 
## eventoy= prueba positiva 
##la probabilidad de x es 0.00148
## la probabilidad de y cuando ya sucedio x es 0.93 p(y|x)
## el complemento de x es 1-p_x, o sea 0.99852
## la probabilidad de y cuando ya sucedio el complemento de x es la probabilidad de que la prueba sea positiva dado que no se tiene la enfermedad, o sea 0.01, pues es 1-0.99
## la probabilidad de y es algo que queremos calcular, hacemos algo similar que lo del problema anterior

p_x=0.00148
p_y_x=0.93
p_y_Xc=0.01
p_Xc=0.99852
p_y=p_y_x*p_x+p_y_Xc*p_Xc

## finalmente 
p_x_y=p_y_x*p_x/p_y
p_x_y

```

```python
#19.Haz un código que escriba letras y el espacio con la misma frecuencia con que aparecen en el cuento de Alicia en el País de las Maravillas

##Lo primero que debemos sacar es la frecuencia con la que se repiten las letras en el cuento. Creo que para mayor simplicidad podemos
## pasar el cuento a minusculas para que no cuente cada mayuscula como algo diferente. 
from collections import Counter
import requests

url = 'https://github.com/andergd/separadorSilabas/blob/master/Carroll%2C%20Lewis%20-%20Alicia%20En%20El%20País%20De%20Las%20Maravillas.txt'  # Reemplaza con la URL real
descarga = requests.get(url)
cuento = descarga.text.lower() 
frecuencia = Counter(cuento)
#Hasta aqui ya tenemos las letras y caracteres especiales contados
#Ahora vamos a juntarlas todas en una secuencia, es decir, como si hicieramos chilaquiles con todas las letras y caracteres
letras = []
for caracter, freq in frecuencia.items():
    letras.extend([caracter] * freq)

import random
#Ahora vamos a generar una funcion que sea la encargada de generar el texto nuevo
def nuevo_cuento(frecuencias, longitud):
    return ''.join(random.choice(frecuencias) for _ in range(longitud)) #como tal lo que hace es juntar letras y caracteres y separarlos
    #justo lo que buscamos

#vamos a ponerle con respecto al tamaño del cuento, aunque podemos cambiar la variable y ponerle menos para que sea mas pequeño
longitud_texto= len(cuento)

texto = nuevo_cuento(letras, longitud_texto)
texto
```
